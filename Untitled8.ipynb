```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
dataset=pd.read_csv('/content/decision_tree.csv')
x=dataset.iloc[:,:-2].values
y=dataset.iloc[:,-1].values

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.25,random_state=0)
print(x_test)
print(x_train)
print(y_test)
print(y_train)

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)

from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier.fit(x_train,y_train)
print(classifier.fit(x_train,y_train))

classifier.predict(sc.transform([[40,60000]]))
print(classifier.predict(sc.transform([[40,60000]])))

y_pred=classifier.predict(x_test)
np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1)) # Corrected np.concatenate syntax

from sklearn.metrics import confusion_matrix,accuracy_score
print("CONFUSION MATRIX:",confusion_matrix(y_test,y_pred))
print("\nACCURACY SCORE:",accuracy_score(y_test,y_pred))

from matplotlib.colors import ListedColormap
x_set,y_set=x_train,y_train
x1,x2=np.meshgrid(
    np.arange(start=x_set[:,0].min()-1,stop=x_set[:,0].max()+1,step=0.025),
    np.arange(start=x_set[:,1].min()-1,stop=x_set[:,1].max()+1,step=0.025)
)
plt.contourf(x1,x2,classifier.predict(np.array([x1.ravel(),x2.ravel(),]).T).reshape(x1.shape),alpha=0.75,cmap=ListedColormap(("lightcoral","lightgreen")))
plt.scatter(x_set[:,0],x_set[:,1],c=y_set,cmap=ListedColormap(('red','green')))
plt.title("Decision Tree Training set")
plt.xlabel("Age")
plt.ylabel("Estimated salary(scaled)")
plt.show()

from matplotlib.colors import ListedColormap
x_set,y_set=x_test,y_test
x1,x2=np.meshgrid(
    np.arange(start=x_set[:,0].min()-1,stop=x_set[:,0].max()+1,step=0.025),
    np.arange(start=x_set[:,1].min()-1,stop=x_set[:,1].max()+1,step=0.025)
)


plt.contourf(x1,x2,classifier.predict(np.array([x1.ravel(),x2.ravel(),]).T).reshape(x1.shape),alpha=0.75,cmap=ListedColormap(("lightcoral","lightgreen")))
plt.scatter(x_set[:,0],x_set[:,1],c=y_set,cmap=ListedColormap(('red','green')))
plt.title("Decision Tree Testing set")
plt.xlabel("Age")
plt.ylabel("Estimated salary(scaled)")
plt.show()
```
